```python
import json
records = {
    1: {"Name": "Redmi Note 7 Pro", "Price": 15000, "Quantity": 10},
    2: {"Name": "Samsung Galaxy M14", "Price": 13000, "Quantity": 8},
    3: {"Name": "iPhone SE", "Price": 30000, "Quantity": 5},
    4: {"Name": "Realme Narzo 60", "Price": 17000, "Quantity": 12},
    5: {"Name": "OnePlus Nord CE", "Price": 25000, "Quantity": 6},
    6: {"Name": "Vivo T1", "Price": 16000, "Quantity": 9},
    7: {"Name": "Poco X5", "Price": 18000, "Quantity": 7},
    8: {"Name": "Motorola G54", "Price": 14000, "Quantity": 11},
    9: {"Name": "Oppo A78", "Price": 19000, "Quantity": 4},
    10: {"Name": "Redmi 13C", "Price": 9000, "Quantity": 15}
}

print("---------------------------------------------------  Product Details ---------------------------------------------\n")

for key,v in records.items():
    print(f"Product Id : {key} | Product Name : {records[key]['Name']} | Product Price : {records[key]['Price']} | Product Quantity Avaiable :{records[key]['Quantity']}")

print("\n-------------------------------------------------------------------------------------------------------------------\n")

ui_pr_id = int(input("Enter the Product Id : "))
ui_pr_qn = int(input("Enter number of quantity required : "))

cal_bill = records[ui_pr_id]["Price"]*ui_pr_qn
records[ui_pr_id]["Quantity"] = records[ui_pr_id]["Quantity"]-ui_pr_qn
print("-----------------  Your Purchase -------------------------")
print("Product Name       : ",records[ui_pr_id]["Name"])
print("Product Price      : ",records[ui_pr_id]["Price"])
print("Product Quantity   : ",ui_pr_qn)
print("----------------------------------------------------------")
print("Bill Amount        : ",cal_bill)
print("\n--------------------------------------------------------\n")

js = json.dumps(records)
fd = open("Inventory.json","w")
fd.write(js)
fd.close()
```
