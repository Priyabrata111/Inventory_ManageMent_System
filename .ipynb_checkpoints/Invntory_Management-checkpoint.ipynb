## Product Details
1. Product Id
2. Product Name
3. Product Price
4. Product Quantity

```python
fd = open("Inventory.txt","r")
txt = fd.read()
fd.close()
```

```python
products = txt.split("\n")
```

```python
products
```

```python
ui_prod_id = int(input("Enter The Product ID (in Numbers)                    : "))
ui_prod_qn = int(input("Enter the Quanity user want to purchase (in Numbers) : "))
prodFound = False
modified_list = []
for product in products:
    prod_details = product.split(',')
    if int(prod_details[0]) == ui_prod_id:
       # print(prod_details[0], prod_details[1],prod_details[2],prod_details[3])
        prodFound = True
        prod_name = prod_details[1]
        prod_price = int(prod_details[2])
        prod_qn = int(prod_details[3])
        print("------------------------------------------")
        print(f"Product Name              : {prod_name}")
        print(f"Product Price             : {prod_price}")
        print(f"Product Avaiable Quantity : {prod_qn}")
        print("-------------------------------------------")
        if prod_qn<ui_prod_qn:
            print(f"Sorry We have only {prod_qn} in our store but the requested quanity is {ui_prod_qn}")
            #print("Please try again after reducing the quantity or wait for stock to be restored.")
            print(f"Would you like to purchase the remaining {prod_qn} from our store ?")
            ch = input("Please Enter Y or N :")
            if ch=='Y' or ch =='y':
                cal_bill = prod_qn*prod_price
                print(f"Bill generated            : {cal_bill}")
                prod_qn = prod_qn-prod_qn
                prod_details[3] = str(prod_qn)
            else:
                print("Thanks!")
                
        else:
            cal_bill = ui_prod_qn*prod_price
            print(f"Bill generated            : {cal_bill}")
            prod_qn = prod_qn-ui_prod_qn
            prod_details[3] = str(prod_qn)
   # print(prod_details)
    modified_list.append(prod_details)
if prodFound == False:
    print("No Product Found")
```

```python
fd = open("Inventory.txt","w")
cnt = 0
for i in modified_list:
    cnt = cnt+1
    new_prod = i[0]+','+i[1]+','+i[2]+','+i[3]
    if cnt<len(products):
        new_prod = new_prod + '\n'
    fd.write(new_prod)
fd.close()
```
