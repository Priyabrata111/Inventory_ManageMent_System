```python
import json
import time

fd = open("Inventory.json","r")
js = fd.read()
fd.close()
records = json.loads(js)

ui_name = str(input("Please Enter Your Name"))
ui_email = str(input("Please Enter Your Email"))
ui_mobile_no = str(input("Please Enter Mobile Number"))

print("---------------------------------------------------  Product Details ---------------------------------------------\n")

for key,v in records.items():
    print(f"Product Id : {key} | Product Name : {records[key]['Name']} | Product Price : {records[key]['Price']} | Product Quantity Avaiable :{records[key]['Quantity']}")

print("\n-------------------------------------------------------------------------------------------------------------------\n")

ui_pr_id = str(input("Enter the Product Id : "))
ui_pr_qn = int(input("Enter number of quantity required : "))


trsc = ''
if ui_pr_qn<=records[ui_pr_id]["Quantity"]:
    cal_bill = records[ui_pr_id]["Price"]*ui_pr_qn
    discount = 0
    if cal_bill<9999:
        discount = 0.1
    elif cal_bill>=9999 and cal_bill<39999:
        discount = 0.2
    else:
        discount = 0.3
        
    records[ui_pr_id]["Quantity"] = records[ui_pr_id]["Quantity"]-ui_pr_qn
    print("-----------------  Your Purchase -------------------------\n")
    print("Product Name       : ",records[ui_pr_id]["Name"])
    print("Product Price      : ",records[ui_pr_id]["Price"])
    print("Product Quantity   : ",ui_pr_qn)
    print("----------------------------------------------------------")
    print("Bill Amount        : ",cal_bill)
    print(f"Discount           : {discount*100}%")
    print("GST                : 9% ")
    cal_bill = cal_bill+0.09*cal_bill-discount*cal_bill
    print("Total Amount       : ",cal_bill)
    print("\n--------------------------------------------------------\n")
    trsc = ui_name+", "+ui_email+", "+ui_mobile_no+", "+records[ui_pr_id]["Name"]+", "+str(records[ui_pr_id]["Price"])+", "+str(ui_pr_qn)+", "+str(cal_bill)+", "+time.ctime()+"\n"
    
else:
    print(f"Sorry We have only {records[ui_pr_id]['Quantity']} quantity but the requested quantity is {ui_pr_qn}")
    print(f"Would you like to purchase {records[ui_pr_id]['Quantity']} quantity ?")
    ch = str(input("Please Enter Y or N : "))
    if ch == 'Y' or ch=='y':
        cal_bill = records[ui_pr_id]["Price"]*records[ui_pr_id]["Quantity"]
        records[ui_pr_id]["Quantity"] = records[ui_pr_id]["Quantity"]-records[ui_pr_id]["Quantity"]
        discount = 0
        if cal_bill<9999:
            discount = 0.1
        elif cal_bill>=9999 and cal_bill<39999:
            discount = 0.2
        else:
            discount = 0.3
        print("-----------------  Your Purchase -------------------------\n")
        print("Product Name       : ",records[ui_pr_id]["Name"])
        print("Product Price      : ",records[ui_pr_id]["Price"])
        print("Product Quantity   : ",records[ui_pr_id]["Quantity"])
        print("----------------------------------------------------------")
        print("Bill Amount        : ",cal_bill)
        print(f"Discount           : {discount*100}%")
        print("GST                : 9% ")
        cal_bill = cal_bill+0.09*cal_bill-discount*cal_bill
        print("Total Amount       : ",cal_bill)
        print("\n--------------------------------------------------------\n")
        trsc = ui_name+", "+ui_email+", "+ui_mobile_no+", "+records[ui_pr_id]["Name"]+", "+str(records[ui_pr_id]["Price"])+", "+str(records[ui_pr_id]["Quantity"])+", "+str(cal_bill)+", "+time.ctime()+"\n"
    else:
        print("Thanks!")
        
    
js = json.dumps(records)
fd = open("Inventory.json","w")
fd.write(js)
fd.close()

if len(trsc)>1:
    fd = open("TransactionsJson.csv","a")
    fd.write(trsc)
    fd.close()
    print("Thanks!!! Please Visit Again !")
```
